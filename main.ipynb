```python

root = 'feb_2023'
tv_path = root+"\\"+"tv.csv"
df = pd.read_csv(tv_path)
```

```python
df.drop('summary', axis=1, inplace=True)
df
```

```python
user_score = df[(df['metascore']>=80) & (df['user_score']!='tbd')].sort_values(['user_score', 'metascore'], ascending=[False, False])
user_score.head(20)
```
